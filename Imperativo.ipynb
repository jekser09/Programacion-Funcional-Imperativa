```python
productos = dict()

def menu():
  print("------Menu------")
  print("1. Añadir producto.")
  print("2. Mostrar el listado.")
  print("3. Consultar producto.")
  print("4. Productos sin stock.")
  print("5. Salir.")
  print("_______________________")
  opcion=input("Digite la opción que desee -> ")
  print("")
  if opcion.isdigit():
    return opcion
  else:
    return 0

def añadirProducto():
  bandera=True
  while bandera:
    try:
      id=int(input("Ingrese el codigo del producto: "))
      if(productos.get(id)==None):
        nombre=input("Ingrese el nombre del producto: ")
        precio=int(input("Ingrese el precio del producto: "))
        stock=int(input("Cantidad de producto: "))
        if(len(nombre)>0 and precio>0 and stock>=0):               
            #validarClave(id)
            productos.setdefault(id,[nombre,precio,stock])
            print("¿Desea agregar más productos? -> ")
            rta=input("¿ Sí / No ?\n-->").upper()
            while True:
              if rta=="SI" or rta=="S":
                print("Continuar...")
                break              
              elif rta=="NO" or rta=="N":
                print("")               
                bandera=False
                break
              else:
                print("Digite una opcion correcta")          
        else:
          print("Por favor complete los datos")
      else:
        print("Este producto ya esta registrado")     
    except ValueError:
      print("La entrada es incorrecta, digite solo numeros.")

def mostrarProductos():
  if(len(productos)==0):
    print("No hay productos registrados")
  else:
    for key in productos:
      print(f"Id: {key}--> {productos[key][0]} con un costo de ${productos[key][1]} y en stock hay {productos[key][2]}\n")   

def buscarProducto(id):
  if productos.get(id) != None:
    print(f"Id {id} --> {productos[id][0]} con un costo de ${productos[id][1]} y en stock hay {productos[id][2]}")
  else:
    print("Este producto no esta registrado.")

def sinStock():
  for key in productos:
    if productos[key][2]==0:
      print(f"{productos[key][0]} stock disponible: {productos[key][2]}\n")

# Opciones
bandera=True
while bandera:
  opcion=int(menu())
  if opcion>0 and opcion<6:
    if opcion==1:
      añadirProducto()
    elif opcion==2:
      mostrarProductos()
    elif opcion==3:
      clave=input("Digite la clave del producto que desea consultar: ")
      buscarProducto(int(clave))
    elif opcion==4:
      sinStock()
    elif opcion==5:
      rta=input("¿ Sí / No ?\n-->").upper()
      if rta=="SI" or rta=="S":
        print("Adios")
        bandera=False
      elif rta=="NO" or rta=="N":
        print("Continuar...")       
  else:
    print("Digite una opcion valida. \n")
```
