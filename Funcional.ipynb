```python
#Definicion diccionario vacio
productos = dict()

#Funcion para que el usuario elija una opcion
def opciones():  
    print("------Menu------")
    print("1. Añadir producto.")
    print("2. Mostrar el listado.")
    print("3. Consultar producto.")
    print("4. Productos sin stock.")
    print("5. Salir.")
    print("_______________________")    
    opcion=input("Digite la opción que desee -> ")
    if opcion.isdigit():
      if int(opcion)>0 and int(opcion)<6:
        return int(opcion)
      else:
        print("Digite una opcion valida")
        return opciones()
    else:
      print("Solo numeros")
      return opciones()

#Funcion que agrega productos al diccionario
def añadirProductos():
  try:
    id=int(input("Ingrese el codigo del producto: "))
    if(productos.get(id)==None):
      nombre=input("Ingrese el nombre del producto: ")
      precio=int(input("Ingrese el precio del producto: "))
      stock=int(input("Cantidad de producto: "))
      if len(nombre)>0 and precio>0 and stock>=0:
        productos.setdefault(id,[nombre,precio,stock])
        print("¿Desea agregar más productos? -> ")
        rta=input("¿ Sí / No ?\n-->").upper()
        if rta=="SI" or rta=="S":
          print("Continuar...")
          añadirProductos()
        elif rta=="NO" or rta=="N":
          programa()
        else:
          print("Opcion incorrecta sera enviado al menu")
          programa()
      else:
        print("Por favor complete los datos")
        añadirProductos()        
    else:
      print("Este producto ya existe")
      añadirProductos()  
  except ValueError:
    print("Revise que los datos sean correctos")
    añadirProductos()

#Recorrer todo el diccionario e imprimir item a item
def mostrarListado(n):  
  if len(productos)!=0:    
    lista=list(productos.keys())    
    if n<len(lista):
      print(f"Id {lista[n]}--> {productos[lista[n]][0]} Costo: {productos[lista[n]][1]} y en stock hay {productos[lista[n]][2]}")
      n=n+1
      mostrarListado(n)
    else:
      programa()
  else:
    print("No hay productos registrados")
    programa()

#Buscar un producto por su id
def consultarProducto():
  try:
    id=int(input("Id del producto que desea buscar"))
    if productos.get(id) != None:
      print(f"Id {id} --> {productos[id][0]} con un costo de ${productos[id][1]} y en stock hay {productos[id][2]}")
      programa()
    else:
      print("Este producto no esta registrado.")
      programa()
  except ValueError:
    print("Solo numeros")
    consultarProducto()

#Busca un producto con 0 en stock
def productosSinStock(nostock,n):
  if len(nostock)!=0:
    if n<len(nostock):
      print(f"Hay {productos[nostock[n]][2]}  {productos[nostock[n]][0]} en el inventario")
      n=n+1
      productosSinStock(nostock,n)
    else:
      programa()
  else:
    print("No hay productos sin stock")
    programa()

#Salir  
def salir():
  return print("salir")

#Estructura del menu
def programa():
  op=opciones()
  if op==1:
    añadirProductos()  
  elif op==2:
    mostrarListado(0)
  elif op==3:
    consultarProducto()
  elif op==4:
    nostock=list(filter(lambda x:productos[x][2]==0,productos))  
    productosSinStock(nostock,0)
  elif op==5:
    return salir()

#Llamado de la funcion programa que es la que controla el
#comportamiento del programa
programa()
```
